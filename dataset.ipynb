```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leitura do arquivo CSV e seleção das colunas de interesse
df = pd.read_csv('datasets/Original_data_with_more_rows.csv')
newdf = df[['ParentEduc', 'MathScore', 'ReadingScore', 'WritingScore']]

# Agrupamento por nível de educação dos pais e cálculo da mediana
newdfGroup = newdf.groupby(['ParentEduc']).median()
print (newdfGroup.head())
```

```python
# Plotagem do gráfico de barras

colors = ['blue','green','yellow']
newdfGroup.plot(kind='bar',color=colors, figsize= (10,6))
plt.xlabel('Nível de Educação dos Pais')
plt.ylabel('Mediana')
plt.title('Escores de Matemática, Leitura e Escrita por Nível de Educação dos Pais')
plt.legend(['Matemática', 'Leitura', 'Escrita'])
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotagem do gráfico de violinplot
plt.figure(figsize=(10, 6))
plt.violinplot(dataset=[newdf[newdf['ParentEduc'] == level]['MathScore'] for level in newdf['ParentEduc'].unique()],
               showmedians=True)
plt.xlabel('Nível de Educação dos Pais')
plt.ylabel('Escores de Matemática')
plt.title('Distribuição dos Escores de Matemática por Nível de Educação dos Pais')
plt.xticks(ticks=range(1, len(newdf['ParentEduc'].unique()) + 1),
           labels=newdf['ParentEduc'].unique(), rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotagem do gráfico de violinplot
plt.figure(figsize=(10, 6))
plt.violinplot(dataset=[newdf[newdf['ParentEduc'] == level]['ReadingScore'] for level in newdf['ParentEduc'].unique()],
               showmedians=True)
plt.xlabel('Nível de Educação dos Pais')
plt.ylabel('Escores de Leitura')
plt.title('Distribuição dos Escores de Leitura por Nível de Educação dos Pais')
plt.xticks(ticks=range(1, len(newdf['ParentEduc'].unique()) + 1),
           labels=newdf['ParentEduc'].unique(), rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotagem do gráfico de violinplot
plt.figure(figsize=(10, 6))
plt.violinplot(dataset=[newdf[newdf['ParentEduc'] == level]['WritingScore'] for level in newdf['ParentEduc'].unique()],
               showmedians=True)
plt.xlabel('Nível de Educação dos Pais')
plt.ylabel('Escores de Redação')
plt.title('Distribuição dos Escores de Redação por Nível de Educação dos Pais')
plt.xticks(ticks=range(1, len(newdf['ParentEduc'].unique()) + 1),
           labels=newdf['ParentEduc'].unique(), rotation=45)
plt.tight_layout()
plt.show()
```
